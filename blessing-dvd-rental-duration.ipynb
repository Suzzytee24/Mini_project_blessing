```python
# Loading the dataset
df = pd.read_csv('/kaggle/input/datasets/blessingtosin/rental-info/rental_info.csv')
```

```python
# A quick preview into the dataset
df.head()
```

```python
# Checking to know more about the dataset
df.info()
```

```python
# Checking for missing values
df.isnull().sum()
```

```python
# In the absence of no missing value
# Coverting the selected data to datetime format using "pd.to_dataset"
df['rental_date'] = pd.to_datetime(df['rental_date'])
df['return_date'] = pd.to_datetime(df['return_date']) 
df.dtypes
```

```python
# to create the terget variable by substracting rental_date form return_date, then using ".dt.days" to convert it to number of days
df['rental_length_days'] = (df['return_date'] - df['rental_date']).dt.days
df['rental_length_days'].head()
```

```python
# coverting text data into numerical features
df['deleted_scenes'] = df['special_features'].str.contains('Deleted Scenes').astype(int)
df['behind_the_scenes'] = df['special_features'].str.contains('Behind the Scenes').astype(int)

print("DELETED SCENES")
print(df['deleted_scenes']) 
print("="*10)
print("BEHIND THE SCENES")
print(df['behind_the_scenes'])
```

```python
# Splitting (defining) data into X (features) and y (target)
X = df.drop(columns=['return_date', 'rental_date', 'special_features', 'rental_length_days'])
y = df['rental_length_days'] 
print("Done")
```

```python
X.head()
```

```python
y.head()
```

```python
# Splitting the data into train and test with a test_size of 20% and random state of 9
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=9) 

print("Data splited Successefully")
```

```python
X_test.head()
```

```python
linear_regression = LinearRegression() 
ridge_regression = Ridge(random_state=9) 
lasso_regression = Lasso(random_state=9)
random_forest = RandomForestRegressor(random_state=9) 

print("Done")
```

```python
linear_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('model', linear_regression)
]) 
print("Done")
```

```python
ridge_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('model', ridge_regression)
])
print("Done")
```

```python
lasso_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('model', lasso_regression)
]) 
print("Done")
```

```python
random_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('model', random_forest)
])
print("Done")
```

```python
ridge_params = {
    'model__alpha': [0.1, 1.0, 5.0, 10.0]
}

lasso_params = {
    'model__alpha': [0.01, 0.05, 0.1, 0.5]
}

random_params = {
    'model__n_estimators': [100, 200],
    'model__max_depth': [10, 20, None],
    'model__min_samples_split': [2, 5],
    'model__min_samples_leaf': [1, 2]
}
print("Done")
```

```python
models = {
    'Linear Regression': (linear_pipeline),
    'Ridge Regression': (ridge_pipeline, ridge_params),
    'Lasso Regression': (random_pipeline, random_params)
} 
print(models)
```

```python
best_trained_models = {} 
results = {}
print("Done")
```

```python
# Applying  linear_pipeline and evaluating the performance usibg mean squared error
linear_pipeline.fit(X_train, y_train) 
y_pred_linear = linear_pipeline.predict(X_test)
mse_linear = mean_squared_error(y_test, y_pred_linear) 
best_trained_models['Linear Regression'] = linear_pipeline
results['Linear Regression'] = mse_linear
print(f"Linear MSE: {mse_linear:.4f}")
```

```python
# Applying Ridge Regression Pipeline with GridSearchCV
# Using GridSearchCV to find the best parameter value to use
ridge_grid = GridSearchCV(
    ridge_pipeline, ridge_params, cv=5,
    scoring='neg_mean_squared_error', n_jobs=-1
)
ridge_grid.fit(X_train, y_train) 
ridge = ridge_grid.best_params_
best_ridge = ridge_grid.best_estimator_

print(f"Best Ridge Parameters: {ridge}")
```

```python
# Evaluating the tuned Ridge using mean_squared_error (round up to 4 decimal place)

y_pred_ridge = best_ridge.predict(X_test) 
mse_ridge = mean_squared_error(y_test, y_pred_ridge) 

results['Ridge Regression'] = mse_ridge
best_trained_models['Ridge Regression'] = best_ridge
print(f"Ridge MSE: {mse_ridge:.4f}")
```

```python
# Applying Lasso Regression Pipeline with GridSearchCV
# Using GridSearchCV to find the best parameter value

lasso_grid = GridSearchCV(
    lasso_pipeline, lasso_params, cv=5, scoring='neg_mean_squared_error', n_jobs=-1
)
lasso_grid.fit(X_train, y_train) 
lasso = lasso_grid.best_params_
best_lasso = lasso_grid.best_estimator_ 

print(f"Best Lasso Parameters: {lasso}")
```

```python
# Evaluating the tuned Lasso using mean_squared_error (round up to 4 decimal place)

y_pred_lasso = best_lasso.predict(X_test) 
mse_lasso = mean_squared_error(y_test, y_pred_lasso) 

best_trained_models['Lasso Regression'] = best_lasso
results['Lasso Regression'] = mse_lasso 

print(f"Lasso: {mse_lasso:.4f}")
```

```python
# Applying Random Forest Regression Pipeline with GridSearchCV
# Using GridSearchCV to find the best parameter value

random_grid = GridSearchCV(
    random_pipeline, random_params, cv=5, scoring='neg_mean_squared_error', n_jobs=-1
)

random_grid.fit(X_train, y_train) 
random = random_grid.best_params_
best_random = random_grid.best_estimator_ 

print(f"Best Random Parameters: {random}")
```

```python
# Evaluating the tuned Random Forest using mean_squared_error (round up to 4 decimal place)

y_pred_random = best_random.predict(X_test) 
mse_random = mean_squared_error(y_test, y_pred_random) 

best_trained_models['Random Forest'] = best_random
results['Random Forest'] = mse_random
print(f"Random MSE: {mse_random:.4f}")
```

```python
# Selecting the Best Model 

best_model_name = min(results, key=results.get) 
best_model = best_trained_models[best_model_name] 
best_mse = results[best_model_name]

print("Done")
```

```python
# Printing the result for the best model

print(f"BEST MODEL: {best_model_name}") 
print(f"MSE: {best_mse:.4f}") 
print(f"Requirement met the (MSE <= 3.0): {best_mse <= 3.0}")
```
